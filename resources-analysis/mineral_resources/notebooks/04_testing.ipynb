```python
import json
import os
import logging
import pandas as pd
```

```python
activities = ["aggregated_exports", "aggregated_imports", "aggregated_production"]
commodities = []
issue_sheets = ["exports_cobalt_1980_1989.xlsx", "exports_cobalt_1990_1999.xlsx", "exports_cobalt_2000_20009.xlsx"]
issue_datasets = {}
#logger = logging.getLogger('ftpuploader')
```

```python
for activity in activities:
    for sheet in os.listdir(f"./{activity}/"):
        path = f"./{activity}/{sheet}"
        files_to_ignore = [".DS_Store", ".ipynb_checkpoints"]
        if any(file in path for file in files_to_ignore):
            print(f"Ignoring <{path}>\n\n")
            continue
        print(f"Trying to open dataset <{path}>...")
        df = pd.read_csv(path, header=0)#.drop(columns=["Unnamed: 0"])
        try:
            df["Amount"] = df["Amount"].astype(float)
            if len(df.columns) != 7:
                raise Exception(f"# of columns in {path} not equal to 7. Investigate.")
            print("No issues detected.")
            print("\n\n")
            
        except Exception as e:
            amount_values = df["Amount"].values()
            print(f"WARNING! {path}")
            print("\n\n")
            issue_datasets[path] = repr(e)
            #df["Amount"] = df["Amount"].astype(float)
```
