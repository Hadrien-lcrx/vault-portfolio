```python
with open('measures_exports.json') as json_file:
    measures_exports = json.load(json_file)
    
with open('measures_imports.json') as json_file:
    measures_imports = json.load(json_file)
    
with open('measures_production.json') as json_file:
    measures_production = json.load(json_file)
    
```

```python
clean_measures_exports = {}
clean_measures_imports = {}
clean_measures_production = {}
```

```python
for k, v in measures_exports.items():
    for sub_k, sub_v in v.items():
        clean_key = re.split('\d+-', sub_k)[1].replace(" ", "_").replace(",", "")
        unique_measures = []
        for year, measure in sub_v.items():
            unique_measures.append(measure)
        unique_measures = set(unique_measures)
        if "missing" in unique_measures:
            unique_measures.remove("missing")
        if len(unique_measures) == 0:
            clean_measures_exports[clean_key] = np.NaN
        else:
            clean_measures_exports[clean_key] = list(unique_measures)[0]
            
for k, v in measures_imports.items():
    for sub_k, sub_v in v.items():
        clean_key = re.split('\d+-', sub_k)[1].replace(" ", "_").replace(",", "")
        unique_measures = []
        for year, measure in sub_v.items():
            unique_measures.append(measure)
        unique_measures = set(unique_measures)
        if "missing" in unique_measures:
            unique_measures.remove("missing")
        if len(unique_measures) == 0:
            clean_measures_imports[clean_key] = np.NaN
        else:
            clean_measures_imports[clean_key] = list(unique_measures)[0]
            
for k, v in measures_production.items():
    for sub_k, sub_v in v.items():
        clean_key = re.split('\d+-', sub_k)[1].replace(" ", "_").replace(",", "")
        unique_measures = []
        for year, measure in sub_v.items():
            unique_measures.append(measure)
        unique_measures = set(unique_measures)
        if "missing" in unique_measures:
            unique_measures.remove("missing")
        if len(unique_measures) == 0:
            clean_measures_production[clean_key] = np.NaN
        else:
            clean_measures_production[clean_key] = list(unique_measures)[0]
```

```python
for k, v in clean_measures_exports.items():
    imports = clean_measures_imports.get(k, f"{k} Not in imports dictionary")
    production = clean_measures_production.get(k, f"{k} Not in production dictionary")
    if (v == imports == production) and (pd.notnull(v)):
        pass
        #print("All good")
    else:
        print(v, imports, production)
```

```python
for folder in cleaned_folders:
    files = [file for file in listdir(f"{path}/{folder}")]
    datasets = [file for file in files if ".csv" in file]
    for mineral in minerals:
        mineral_datasets = [file for file in files if (".csv" in file) and (re.search(f"^{mineral}\_\d*", file))]
        if len(mineral_datasets) > 0:
            df = pd.DataFrame(columns=["Country", "Activity", "Commodity", "Sub-commodity"])
            
            # Fill in existing data
            for dataset in mineral_datasets:
                df_to_join = pd.read_csv(f"{path}/{folder}/{dataset}", header=0)
                df_to_join = df_to_join.astype({'Sub-commodity': 'str'})
                df = pd.merge(df, df_to_join, how="outer",
                              left_on=["Country", "Activity", "Commodity", "Sub-commodity"],
                              right_on=["Country", "Activity", "Commodity", "Sub-commodity"], suffixes=("", "_y"))
                df = df.drop(df.filter(regex='_y$').columns, axis=1)
                df = df.loc[:,~df.columns.duplicated()].copy()
                
                subcommodities = df["Sub-commodity"].unique()
            
            # Add NaN countries
            #for country in countries:
            #    if country not in df["Country"].values:
            #        if len(subcommodities) == 1:
            #            for subcommodity in subcommodities:
            #                existing_values = [country, mineral]
            #                missing_values = [np.nan] * (len(df.columns) - len(existing_values))
            #                row = existing_values + missing_values
            #                df.loc[len(df)] = row
            #        else:
            #            for subcommodity in subcommodities:
            #                if pd.notnull(subcommodity):
            #                    existing_values = [country, mineral, subcommodity]
            #                    missing_values = [np.nan] * (len(df.columns) - len(existing_values))
            #                    row = existing_values + missing_values
            #                    df.loc[len(df)] = row
                        
            # Add NaN columns
            #for year in years:
            #    if year not in df.columns:
            #        df[year] = np.nan
            #reordered_columns = ["Country", "Commodity", "Sub-commodity", *years]
            #df = df[reordered_columns]
            #df = df.sort_values(by="Country", axis=0)
            if "exports" in folder:
                for k in clean_measures_exports.keys():
                    if mineral in k:
                        df["Measure"] = clean_measures_exports[k]
                        break
                    else:
                        continue
            elif "imports" in folder:
                for k in clean_measures_imports.keys():
                    if mineral in k:
                        df["Measure"] = clean_measures_imports[k]
                        break
                    else:
                        continue
            elif "production" in folder:
                for k in clean_measures_production.keys():
                    if mineral in k:
                        df["Measure"] = clean_measures_production[k]
                        break
                    else:
                        continue
            # Move years as column
            #print(f"Trying to melt {mineral}...")
            df = df.melt(id_vars=["Country", "Activity", "Commodity", "Sub-commodity", "Measure"], var_name="Year", value_name="Amount")
            save_name = f"{folder.split('_')[1]}_{mineral}"
            print(f"Saved {save_name}")
            df.to_csv(f"/app/mineral_resources/aggregated_{folder.split('_')[1]}/{save_name}.csv", header=True, index=False)
            #print(f"{mineral}: {save_name} saved ({len(df)})") 
                

            
```
